```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criação do DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
# Convertendo o dicionário em uma lista para usar como plot
colunas_2022 = list(vendas_2022.keys())

# Configurando a figure para usar 4 gráficos (2,2) e alterando o tamanho
fig, axs = plt.subplots(2,2, figsize=(16, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Tendência de vendas em cada loja', x=0.01, ha='left', fontweight='bold', fontsize=18)

# Plot de cada loja individual
axs[0,0].plot(df.loc['A', colunas_2022], lw=3, color='red')
axs[0,0].set_title('Loja A', loc='left', fontsize=14)
axs[0,0].grid()

axs[0,1].plot(df.loc['B', colunas_2022], lw=3, color='green')
axs[0,1].set_title('Loja B', loc='left', fontsize=14)
axs[0,1].grid()

axs[1,0].plot(df.loc['C', colunas_2022], lw=3, color='teal')
axs[1,0].set_title('Loja C', loc='left', fontsize=14)
axs[1,0].grid()

axs[1,1].plot(df.loc['D', colunas_2022], lw=3, color='blue')
axs[1,1].set_title('Loja D', loc='left', fontsize=14)
axs[1,1].grid()

# Adicionando labels no eixo X e Y em cada gráfico utilizando o for
for ax in axs.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Tendência de vendas')
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)
    
ymin = 0
ymax = 450

# Setando o range minimo e maximo do eixo Y
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
    
fig.savefig('atividade_tendencia_vendas.png', transparent=False, dpi=300, bbox_inches='tight')
  
plt.show()
```

```python
# Usei um for para reduzir o código
fig, axs = plt.subplots(2,2, figsize=(16, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Tendência de vendas em cada loja', x=0.01, ha='left', fontweight='bold', fontsize=18)

# Lista de lojas e cores
lojas = ['A', 'B', 'C', 'D']
cores = ['red', 'green', 'teal', 'blue']

# Percorrendo todos os subplots, lojas e cores ao mesmo tempo
for ax, loja, cor in zip(axs.flat, lojas, cores):
    ax.plot(df.loc[loja, colunas_2022], lw=3, color=cor)
    ax.set_title(f'Loja {loja}', loc='left', fontsize=14)
    ax.set_xlabel('Mês')
    ax.set_ylabel('Tendência de vendas')
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.set_ylim(0, 450)
    ax.grid(True)
    
plt.show()
```
